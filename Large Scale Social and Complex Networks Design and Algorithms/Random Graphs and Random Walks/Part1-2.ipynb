```r
library('igraph')
# 2 (a)
g_barabasi <- barabasi.game(1000, m = 1, directed=F)
cat("The Barabasi undirected network is connected? ", is.connected(g_barabasi))
```

```r
# 2 (b)
fc <- cluster_fast_greedy(g_barabasi)
cat("The modularity of the network is:", modularity(fc), "\n")
plot(fc,g_barabasi,vertex.size=5, vertex.label=NA)
cmsize <- sizes(fc)
print(cmsize)
```

```r
# 2. (c)
g_barabasi_larger <- barabasi.game(10000, m = 1, directed=F)
fc_larger <- cluster_fast_greedy(g_barabasi_larger)
cat("The modularity of the larger Barabasi network (n=10000) is:", modularity(fc_larger))
```

```r
# 2. (d)
plot(degree.distribution(g_barabasi),main="Log-log Degree Distribution of n=1000 Barabasi Network",xlab="Degree",ylab="Frequency",log="xy")
plot(degree.distribution(g_barabasi_larger),main="Log-log Degree Distribution of n=10000 Barabasi Network",xlab="Degree",ylab="Frequency",log="xy")

```

```r
# 2.(e)
n <- 1000
degrees <- rep(0, n)
for(i in 1:n){
    v_i <- sample(V(g_barabasi),1) # random selected vertice
    v_js <- neighbors(g_barabasi, v_i) # the neighbour
    v_j <- sample(v_i,1)
    degrees[i] <- degree(g_barabasi, v_j)
}
plot(density(degrees),type="p",main="Log-Log Degree Distribution Plot",xlab="Log(Degree)",ylab="Log(Frequency)",log ='xy')
```

```r
# 2.(f)
n_f <- 1000
degrees_f <- rep(0, n)
for(i in 1:n_f){
    g_f <- barabasi.game(1000, m = 1, directed=F)
    degrees_f <- degrees_f + degree(g_f)
}
degrees_f <- degrees_f / 1000
plot(rev(degrees_f), main="Ages of Nodes vs. Their Expected Degrees (Average of 1000 Times)",xlab="Age",ylab="Expected Degree")
```
